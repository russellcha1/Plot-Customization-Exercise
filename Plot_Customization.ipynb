#### Task 6

```python
ax = df.plot(
    kind='scatter',
    x='age',
    y='charges',
    figsize=(10, 5)
)
plt.title('Patient Age vs. Insurance Charges', fontsize = 16, fontweight = 'bold')
plt.xlabel('Age (years)')
plt.ylabel('Annual Charges (USD)')
plt.ylim(5000, 30000)
plt.show()
```

#### The graph cuts off the whole third band of data and also doesnt show the band is highly correlated. Your not seeing the picture of the data.

#### Most charts should start at 0 if it is positively correlated. In most data x = 0 and y = 0 will be the data points. When not starting at zero if the dependant variable starts at significantly high number. That way you can see the correlation more clearly at y=0.

## Task 7

```python
region_avg = df.groupby('region')['charges'].mean()
ax = region_avg.plot(kind='bar', figsize=(8, 5))
plt.title("Charges By Region")
plt.xlabel('Region')
plt.ylabel('Charges USD')
plt.xticks(rotation=45, ha='right')
plt.show()
```

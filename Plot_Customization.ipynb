### Task 16

```python
plt.figure(figsize=(10, 5))
ax = sns.scatterplot(
    data=df,
    x='age',
    y='charges',
    hue='smoker',
    alpha=0.4,
)
plt.title("charges by age")
plt.xlabel("age")
plt.ylabel("charges")

avg_charges = df["charges"].mean()
print(avg_charges)

plt.axhline(
    y=avg_charges,
    color='red',
    linestyle='--',
    linewidth=1.5
)

plt.axvline(
    x=40,
    color='blue',
    linestyle=':',
    linewidth=1.0
)
plt.show()
```

1. What do the two lines help the reader notice?
A: the mean charge and the 40 age group.
splits into four quartiles so you can see the demographic in each one. Once hue was set to "smoker" it really shows distinc data sets.

### Task 17

```python
plt.figure(figsize=(10, 5))
ax = sns.scatterplot(
    data=df,
    x='age',
    y='charges',
    alpha=0.4,
)
plt.title("charges by age")
plt.xlabel("age")
plt.ylabel("charges")

avg_charges = df["charges"].mean()
print(avg_charges)

plt.axhspan(
    ymin=20000,
    ymax=50000,
    alpha=0.1,
    color='yellow'
)

plt.axvspan(
    xmin=50,
    xmax=64,
    alpha=0.1,
    color='blue'
)

plt.show()
```

the two bands tell the reader which datapoints fall into specific ranges on the axis that band is on. yellow band-charges(20k-50k), blue band-age(50-60).

### Task 18

```python
max_row = df.loc[df['charges'].idxmax()]
print(max_row)
max_row_age = max_row['age']
max_row_charge = max_row['charges']
```

```python
plt.figure(figsize=(10, 5))
ax = sns.scatterplot(
    data=df,
    x='age',
    y='charges',
    alpha=0.4,
)
plt.title("Charges by Age")
plt.xlabel("Age in Years")
plt.ylabel("Charges in USD")
plt.annotate('Highest Charge',
             xy=(max_row_age, max_row_charge),
             xytext=(max_row_age + 5, max_row_charge - 5000),
             arrowprops=dict(arrowstyle='->', color='black')
)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
ax = sns.scatterplot(
    data=df,
    x='age',
    y='charges',
    alpha=0.4,
)
plt.title("Charges by Age")
plt.xlabel("Age in Years")
plt.ylabel("Charges in USD")
plt.annotate('Highest Charge',
             xy=(max_row_age, max_row_charge),
             xytext=(max_row_age + 5, max_row_charge - 5000),
             arrowprops=dict(arrowstyle='->', color='black')
)
plt.axhline(
    y=avg_charges,
    color='red',
    linestyle='--',
    linewidth=1.5
)

plt.axvline(
    x=40,
    color='blue',
    linestyle=':',
    linewidth=1.0
)
plt.text(x=25,
         y=avg_charges-1500,
         s='Average charges',
         fontsize=9,
         color='black')
plt.show()
```

```python
region_avg = df.groupby('region')['charges'].mean()
```

```python
max_region = region_avg.idxmin()
colors = ['red' if x == max_region else 'gray' for x in region_avg.index]
ax = region_avg.plot(kind='bar',
                     figsize=(8,5),
                     color=colors)
plt.title("Charges By Region")
plt.xlabel('Region')
plt.ylabel('Charges USD')
plt.xticks(rotation=45, ha='right')
plt.show()
```

1. Which region is the lowest-cost?
A: Southwest

2. What is the reader supposed to notice first in your chart? Did it work?
A: The reader is supposed to see the highlighted bar, and it does work because it stands out

```python
mean_BMI = df['bmi'].mean()
mean_charges = df['charges'].mean()
```

```python
plt.figure(figsize=(11,6))
ax = sns.scatterplot(
    data = df,
    x='bmi',
    y='charges',
    hue='smoker',
    alpha=.45,
    palette='colorblind'
)
plt.title('Charges by BMI and Smoking Status', size=15, fontweight='bold')
plt.xlabel('BMI', size=12)
plt.ylabel('Charges (USD)', size=12)
plt.axvline(x=mean_BMI, color='black')
plt.axhline(y=mean_charges)
plt.text(x = 48, y= mean_charges+1000, s='Average Charges', color='red')
plt.text(x=26, y=55000, s='Average BMI', color='black')
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.grid(axis='y', linestyle='--', linewidth=0.5, color='gray')
plt.legend(loc='upper left', title='Smoker')
plt.savefig('bmi_vs_charges_polished.png',
            dpi=150)
plt.show()
```
